### The percentage distribution of the data in each year using the pie chart

```python
df['year'].value_counts().plot(kind='pie',autopct = '%1.1f%%',figsize=(15,6))
```

### The distribution of the variable "price per lb" using displot.

```python
plt.figure(figsize=(15,6))
sns.displot(data=df,x='priceperlb')
```

1. We can interpret from the above plot that skew toward right.
2. The range between 1.4 to 1.5 has the large number of observations.
3. And range between 3.9 to 4.0 has the smaller number of observations.

### The relationship between the variables numcol and prodval through scatterplot

```python
sns.scatterplot(df,x='numcol',y='prodvalue')
```

1. Here we observed that as the Number of honey producing colonies is increases the Value of production is also increases.
2. Most of the numcol are in range between 0 to 100000
3. The highest no of honey producing colonies is above 500000.

### The relationship between categorical variable year and a numerical variable prodvalue through boxplot

```python
sns.boxplot(df,x='year')
```

1. Here 2005 is the median year.
2. There is no outlier is the year.
3. The year starting value is 1998 and the ending value is 2012.

```python
sns.boxplot(df,x='prodvalue')
```

1. The median value of prodvalue is nearly 0.2.
2. There are many outliers in prodvalue.

###  The relationship between the multiple pairs of variables throughout different years using pairplot

```python
sns.pairplot(df,x_vars=['numcol','yieldpercol','totalprod','prodvalue','year'],hue='year')
```

1. The above plot shows the relationship between all these 'numcol', 'yieldpercol', 'totalprod', 'prodvalue','year' variables.
2. The year wise relationship we show in above plot.
